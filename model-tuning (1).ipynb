```python
import pandas as pd
import os
from sklearn import tree 
from sklearn import preprocessing
from sklearn import model_selection
from sklearn import impute
```

```python
dir = "C:/Users/pc/Downloads/ai-level1/titanic"
titanic_train = pd.read_csv(os.path.join(dir, "train.csv"))
titanic_train.shape
titanic_train.info()
```

```python
imp_age = impute.SimpleImputer()
imp_age.fit(titanic_train[['Age']])
titanic_train['Age'] = imp_age.transform(titanic_train[['Age']])
titanic_train.info()
```

```python
le_sex = preprocessing.LabelEncoder()
le_sex.fit(titanic_train['Sex'])
print(le_sex.classes_)
titanic_train['Sex_encoded'] = le_sex.transform(titanic_train['Sex'])
titanic_train.head(20)
```

```python
le_emb = preprocessing.LabelEncoder()
le_emb.fit(titanic_train['Embarked'])
print(le_emb.classes_)
titanic_train['Embarked_encoded'] = le_emb.transform(titanic_train['Embarked'])
titanic_train.head(20)
```

```python
features = ['Parch', 'SibSp', 'Age', 'Sex_encoded', 'Embarked_encoded']
X_train = titanic_train[features]
y_train = titanic_train['Survived']
```

```python
dt = tree.DecisionTreeClassifier()
dt_param_grid = {'max_depth':list(range(1, 11)), 'min_samples_split':[2, 5, 10]}
cv = model_selection.KFold(n_splits=10)
clf = model_selection.GridSearchCV(dt, dt_param_grid, cv=cv, scoring='accuracy',return_train_score=True)
clf.fit(X_train, y_train)
print(clf.best_params_)
print(clf.best_score_)
print(clf.best_index_)
print(clf.best_estimator_)
df = pd.DataFrame(clf.cv_results_)
columns = ['params', 'split0_test_score', 'split1_test_score', 'split9_test_score', 'mean_test_score']
print(df[columns])
```

```python
dt = tree.DecisionTreeClassifier()
dt_param_grid = {'max_depth':list(range(1, 101)), 'min_samples_split':[2, 5, 10], 'criterion':['gini', 'entropy']}
cv = model_selection.KFold(10)
clf = model_selection.GridSearchCV(dt, dt_param_grid, cv=cv, scoring='accuracy',return_train_score=True)
clf.fit(X_train, y_train)
print(clf.best_params_)
print(clf.best_score_)
print(clf.best_index_)
print(clf.best_estimator_)
df = pd.DataFrame(clf.cv_results_)
columns = ['params', 'split0_test_score', 'split1_test_score', 'split9_test_score', 'mean_test_score']
print(df[columns])
```

```python
dt = tree.DecisionTreeClassifier()
dt_param_grid = {'max_depth':list(range(1, 11)), 'min_samples_split':[2, 5, 10], 'criterion':['gini', 'entropy']}
cv = model_selection.KFold(10)
clf = model_selection.RandomizedSearchCV(dt, dt_param_grid, n_iter=100, cv=cv, scoring='accuracy',return_train_score=True)
clf.fit(X_train, y_train)
print(clf.best_params_)
print(clf.best_score_)
print(clf.best_index_)
print(clf.best_estimator_)
print(clf.best_estimator_.score(X_train, y_train))
df = pd.DataFrame(clf.cv_results_)
columns = ['params', 'split0_test_score', 'split1_test_score', 'split9_test_score', 'mean_test_score']
print(df[columns])
```

```python
%matplotlib qt
tree.plot_tree(clf.best_estimator_, feature_names=features, filled=True, fontsize=7)
```

```python
from sklearn import neural_network
nn = neural_network.MLPClassifier()
nn_param_grid = {'hidden_layer_sizes':[(10, 5), (20,10)], 'activation':['logistic', 'tanh', 'relu']}
cv = model_selection.KFold(10)
clf = model_selection.GridSearchCV(nn, nn_param_grid, cv=cv, scoring='accuracy',return_train_score=True)
clf.fit(X_train, y_train)
print(clf.best_params_)
print(clf.best_score_)
print(clf.best_index_)
print(clf.best_estimator_)
df = pd.DataFrame(clf.cv_results_)
columns = ['params', 'split0_test_score', 'split1_test_score', 'split9_test_score', 'mean_test_score']
print(df[columns])
```

```python
titanic_test = pd.read_csv(os.path.join(dir, "test.csv"))
print(titanic_test.shape)
titanic_test.head()
```

```python
titanic_test['Sex_encoded'] = le_sex.transform(titanic_test['Sex'])
titanic_test['Embarked_encoded'] = le_emb.transform(titanic_test['Embarked'])
titanic_test['Age'] = imp_age.transform(titanic_test[['Age']])
X_test = titanic_test[features]
titanic_test['Survived'] = clf.predict(X_test)
titanic_test.head()
```
